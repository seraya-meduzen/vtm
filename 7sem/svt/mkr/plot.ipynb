```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#  f = sin pi x sin pi y

mesh_diameter = [10, 100, 1000]
C_norm = [0.015655, 0.00159129, 0.000159155]
L2_norm = [0.000757632, 7.28105e-06, 7.2399e-08]
h_2 = 1 / np.array(mesh_diameter) ** 2
h = 1 / np.array(mesh_diameter)

fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(20, 10))
axes.set_yscale('log')
axes.set_xscale('log')
axes.plot(mesh_diameter, C_norm, linewidth=2.5, color='m', label='C_norm')
axes.plot(mesh_diameter, L2_norm, linewidth=2.5, color='b', label='L2_norm')
axes.plot(mesh_diameter, h_2, linewidth=2.5, color='g', label='$1~/~h^2$')
axes.plot(mesh_diameter, h, linewidth=2.5, color='black', label='$1 ~/ ~h$')
axes.set_xlabel('Mesh diameter', fontsize=12)
axes.set_ylabel('Approximation', labelpad=13, fontsize=12)
axes.set_facecolor('#f4f4f4')
axes.grid(color='white')
# axes.xaxis.set_ticks(mesh_diameter)
axes.legend(loc="upper left")
```
