```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#  f = sin4x sin 4y

mesh_diameter = [0.152022, 0.0760108, 0.0331463,
                 0.0190031, 0.0165732, 0.00828729, 0.00414365, 0.002188]
C_norm = [0.0107258, 0.00406679, 0.001272,
          0.00042601, 0.000391, 0.000116,  3.4e-05, 9e-06]
L2_norm = [0.0137699, 0.00351263, 0.000629176, 0.000221842,
           0.000157538, 3.94032e-05, 9.85196e-06, 2.91071e-06]
h_2 = np.array(mesh_diameter) ** 2 / 2
h = np.sqrt(h_2)

fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(20, 10))
axes.set_yscale('log')
axes.set_xscale('log')
axes.plot(mesh_diameter, C_norm, linewidth=2.5, color='m', label='C_norm')
axes.plot(mesh_diameter, L2_norm, linewidth=2.5, color='b', label='L2_norm')
axes.plot(mesh_diameter, h_2, linewidth=2.5, color='g', label='$h^2$')
axes.plot(mesh_diameter, h, linewidth=2.5, color='black', label='$h$')
axes.set_xlabel('Mesh diameter', fontsize=12)
axes.set_ylabel('Approximation', labelpad=13, fontsize=12)
axes.set_facecolor('#f4f4f4')
axes.grid(color='white')
# axes.xaxis.set_ticks(mesh_diameter)
axes.legend(loc="upper left")
```
