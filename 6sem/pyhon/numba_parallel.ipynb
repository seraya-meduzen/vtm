```python
def gcd(x, y):
    while (y):
        x, y = y, x % y
    return x

def phi(n):
    cnt = 0
    for i in range(1, n):
        if gcd(i, n) == 1:
            cnt += 1
    return cnt
```

```python
a = np.random.randint(low=1, high=10**4, size=1000)
```

```python
%%time

[phi(n) for n in a]
```

```python
%%time

with Pool(processes=cpu_count()) as pool:
    res = pool.map(phi, a)
```

```python
@numba.jit(nopython=True)
def gcd_numba(x, y):
    while (y):
        x, y = y, x % y
    return x

@numba.vectorize([numba.int64(numba.int64)], nopython=True,
                 target='parallel')
def phi_numba(n):
    cnt = 0
    for i in range(1, n):
        if gcd_numba(i, n) == 1:
            cnt += 1
    return cnt
```

```python
from numba import prange

@numba.jit(nopython=True, parallel=True)
def my_sum(a, b):
    c = np.zeros_like(a)
    for i in prange(a.shape[0]):
        c[i] = (a[i]**2 + b[i]**3)**5
    return c
```

```python
@numba.jit(nopython=True, parallel=True)
def my_sum2(a):
    s = -np.inf
    for i in prange(a.shape[0]):
        s = max(s, a[i])
    return s
```

```python
@numba.jit(nopython=True, parallel=True)
def my_func(x):
    n = x.shape[0]
    
    a = np.zeros_like(x)
    b = np.zeros_like(x)
    s = 0
    
    for i in prange(n):
        a[i] = np.sin(x[i])
    for i in prange(n):
        b[i] = np.cos(a[i]**2)
    for i in prange(n):
        s += b[i]
        
    return s
```

```python
@numba.jit(nopython=True, parallel=True)
def func1():
    n = 100
    vals = np.empty(n)
    for i in prange(n):
        cur = i + 1
        for j in range(i):
            if cur % 2 == 0:
                cur //= 2
            else:
                cur = cur * 3 + 1
        vals[i] = cur
    return vals
```

```python
@numba.jit(nopython=True, parallel=True)
def func1(n):
    acc = 0
    print(get_parallel_chunksize()) # Will print 4.
    for i in prange(n):
        print(get_parallel_chunksize()) # Will print 0.
        acc += i
    print(get_parallel_chunksize()) # Will print 4.
    return acc


@numba.jit(nopython=True, parallel=True)
def func2(n):
    acc = 0
    # This version gets the previous chunksize explicitly.
    old_chunksize = get_parallel_chunksize()
    set_parallel_chunksize(8)
    for i in prange(n):
        acc += i
    set_parallel_chunksize(old_chunksize)
    return acc

# This version saves the previous chunksize as returned
# by set_parallel_chunksize.
old_chunksize = set_parallel_chunksize(4)
result1 = func1(12)
result2 = func2(12)
result3 = func1(12)
set_parallel_chunksize(old_chunksize)
```

```python
import numpy as np
import numba

@numba.jit(nopython=True)
def power_method(A, v):
    u = v.copy()
    for i in range(3 * 10**3):
        u = A @ u
        u /= np.linalg.norm(u)
    return u

@numba.jit(nopython=True, parallel=True)
def iterate_grid_seq(A, scale, sz):
    assert A.shape[0] == A.shape[1] == 3
    n = A.shape[0]

    results = np.empty((sz**3, n))
    tmp = np.linspace(-scale, scale, sz)
    
    for i1 in range(sz):
        v = np.empty(n, dtype=np.float64)
        v1 = tmp[i1]
        for i2, v2 in enumerate(tmp):
            for i3, v3 in enumerate(tmp):
                v[0] = v1
                v[1] = v2
                v[2] = v3

                u = power_method(A, v)

                idx = i1 * sz**2 + i2 * sz + i3
                results[idx] = u.copy()
    
    return results

@numba.jit(nopython=True, parallel=True)
def iterate_grid_par(A, scale, sz):
    assert A.shape[0] == A.shape[1] == 3
    n = A.shape[0]

    results = np.empty((sz**3, n))
    tmp = np.linspace(-scale, scale, sz)
    
    for i1 in numba.prange(sz):
        v = np.empty(n, dtype=np.float64)
        v1 = tmp[i1]
        for i2, v2 in enumerate(tmp):
            for i3, v3 in enumerate(tmp):
                v[0] = v1
                v[1] = v2
                v[2] = v3

                u = power_method(A, v)

                idx = i1 * sz**2 + i2 * sz + i3
                results[idx] = u.copy()
    
    return results

n = 3
np.random.seed(0)
A = np.random.randn(n, n)

res_seq = iterate_grid_seq(A, 5.0, 20)
res_par = iterate_grid_par(A, 5.0, 20)

print(res_seq.sum())
print(res_par.sum())
print(np.abs(res_seq - res_par).sum())
```
